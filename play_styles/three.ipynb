```python
sel['fly']=sel.prgp/sel.prgc
sel['speed']=(sel.poss*90)/sel.prgdist
sel
```

```python
import matplotlib.pyplot as plt
x = sel['total att']
y = sel['speed']

import seaborn as sns

ax = sns.lmplot(data=sel,x='total att',y='speed',fit_reg=False)

plt.title('Example Plot')
plt.xlabel('total passes')
plt.ylabel('speed')


def label_point(x, y, val, ax):
    a = pd.concat({'x': x, 'y': y, 'val': val}, axis=1)
    for i, point in a.iterrows():
        ax.text(point['x']+.02, point['y'], str(point['val']))

label_point(x, y,sel['team'], plt.gca())  

plt.axvline(x.mean(), c='orange', ls='--', lw=1.5)
plt.axhline(y.mean(), c='orange', ls='--', lw=1.5)
```
