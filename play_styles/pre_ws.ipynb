```python

def starting11(sel):
    step=sel.sort_values(['minute','second']).groupby(['teamid','playerid'],sort=False).count().reset_index()
    step=step.groupby(['teamid']).head(11)
    return step['playerid'].values

def absw(col):
    return np.mean(np.abs(col-50))

def halfswitch(row):
    
    if row.teamid!=162:
    # if row.period=="FirstHalf":
   
        row.loc['x']=100-row['x']
        row.loc['y']=50-row['y']+50
    return row
```

```python
def starting11game(sel):
    step=sel.sort_values(['game_id','minute','second']).groupby(['game_id','teamid','playerid'],sort=False).count().reset_index()
    step=step.groupby(['game_id','teamid']).head(11)
    step['starting']=True
    return step[['game_id','playerid','starting']]
```

```python
sel=events[selcol]

dfl=sel.set_index(['game_id','playerid'])
dfr=starting11game(sel)
dfr=dfr.set_index(['game_id','playerid'])
dft=dfl.join(dfr)
dft=dft[dft['starting']==True]
```
