```python
import soccerdata as sd
# https://fbref.com/en/comps/9/Premier-League-Stats
team_match_stats = ["schedule","shooting","keeper","passing","passing_types","goal_shot_creation","defense","possession","misc"]
seasons=list(range(19,23))
fbref = sd.FBref(leagues=['ENG-Premier League'], seasons=seasons)
```

```python
# https://fbref.com/en/squads/18bb7c10/2023-2024/matchlogs/all_comps/schedule
match_stats0=fbref.read_team_match_stats(stat_type=team_match_stats[0])
match_stats0
```

```python
matches = match_stats0.copy()
```

```python
def matches_prep(matches):
    # matches=matches.reset_index(2)
    # matches=matches.droplevel([0,1])
    matches=matches.droplevel([0])
    matches=matches.reset_index(1)
    matches.columns = [c.lower() for c in matches.columns]
    return matches

matches=matches_prep(matches)

seasons=sorted(set([x[0] for x in matches.index]))
```

```python
for season in seasons:
    s_matches=matches.loc[season]
    s_matches.to_csv(f"matches{season}.csv")
```
