```python
import soccerdata as sd
# https://fbref.com/en/comps/9/Premier-League-Stats
team_stats = ["standard","keeper","keeper_adv","shooting","passing","passing_types","goal_shot_creation","defense","possession","playing_time","misc",]
team_match_stats = ["schedule","shooting","keeper","passing","passing_types","goal_shot_creation","defense","possession","misc"]
# fbref = sd.FBref(leagues=['ENG-Premier League'], seasons=['2324'])
```

```python
fbref = sd.FBref(leagues=['ENG-Premier League'], seasons=['2223'])
season_stats0 = fbref.read_team_season_stats(stat_type=team_stats[0])
season_stats0
```

```python
# https://fbref.com/en/squads/18bb7c10/2023-2024/matchlogs/all_comps/schedule
match_stats0=fbref.read_team_match_stats(stat_type=team_match_stats[0])
match_stats0
```

```python
match_stats1=fbref.read_team_match_stats(stat_type=team_match_stats[1])
match_stats1
```

```python
matches = match_stats0.copy()
shooting = match_stats1.copy()
```

```python
matches=matches.reset_index(2)
matches=matches.droplevel([0,1])

shooting.columns=[col[1] if col[1] else col[0] for col in shooting.columns.values]
shooting =shooting.reset_index(2)
shooting=shooting.droplevel([0,1])
```

```python
# flatten multiindex
# https://stackoverflow.com/questions/14507794/how-to-flatten-a-hierarchical-index-in-columns
# df.columns = [' '.join(col).strip() for col in df.columns.values]
```

```python
match_df=matches.merge(shooting[['team',"date", "Sh", "SoT", "Dist", "FK", "PK", "PKatt"]],on=['team','date'])
match_df.columns = [c.lower() for c in match_df.columns]
```

```python
match_df.to_csv("matches2223.csv")
```
